### See the na distribution

```python
# col_name: ''
def see_dis(col_name):
    var=all_dat_x[col_name] # data type of var is pandas series
    if var.dtype=='O':
        all_dat_x.groupby(col_name).size().plot.pie(title=col_name)
    else:
        plt.hist(var)
        plt.title(col_name)
```

```python
# see distribution first

# 19 col containing NA (exclude 'PoolQC'==> need to be delete)
plt.figure(figsize=(16,60))

for i in range(len(na_col)):
    plt.subplot(11,3,i+1)
    see_dis(na_col[i])

```

### categorical NA imputatuion

```python
# fill cate var with mode
def get_cat_mode(col_name):
    a=all_dat_x.groupby(col_name).size()
    b=np.max(a)
    return(a.index[np.where(a==b)][0])
```

```python
cat_var_list=[]
conti_var_list=[]
a=all_dat_x.columns
for i in range(len(a)):
    if all_dat_x[a[i]].dtype=='O':
        cat_var_list.append(a[i])
    else:
        conti_var_list.append(a[i])
del(a)
    
```

```python
for i in range(len(cat_var_list)):
    all_dat_x[cat_var_list[i]]=all_dat_x[cat_var_list[i]].fillna(get_cat_mode(cat_var_list[i]))

#======> inside the loop
#all_dat_x['MSZoning']=all_dat_x['MSZoning'].fillna(get_cat_mode('MSZoning'))
#all_dat_x['Alley']=all_dat_x['Alley'].fillna(get_cat_mode('Alley')) ======> inside the loop

# check data type: type()
```

### continuous NA imputation

```python
# fill conti var with mode if the distribution is skewed
for i in range(len(conti_var_list)):
    all_dat_x[conti_var_list[i]]=all_dat_x[conti_var_list[i]].fillna(all_dat_x[conti_var_list[i]].mode()[0])



#all_dat_x['LotFrontage']=all_dat_x['LotFrontage'].fillna(all_dat_x['LotFrontage'].mode()[0]) 
#all_dat_x['MasVnrArea']=all_dat_x['MasVnrArea'].fillna(all_dat_x['MasVnrArea'].mode()[0]) 
#all_dat_x['GarageYrBlt']=all_dat_x['GarageYrBlt'].fillna(all_dat_x['GarageYrBlt'].mode()[0]) 
########### must add mode[0]!!!!!!!!!!!!!!!!
#!!!!!!!!!!!!!
```

## Lasso regression

```python
from sklearn.model_selection import GridSearchCV
from sklearn.metrics import mean_squared_error
from sklearn.linear_model import Lasso
```

```python
las_reg=Lasso()
las_grid=GridSearchCV(estimator=las_reg,param_grid={'alpha':[0.001,0.01,0.1,1,10,100]},
                     scoring='neg_root_mean_squared_error',cv=10)
las_grid.fit(X=train_x,y=log_train_y)
```

```python
cv_result_df=pd.DataFrame(las_grid.cv_results_)
```

```python
cv_result_df
```

Accodring to the lasso result, we can see as the penality hyperparameter lambda increase, the perforamce decrease. The potential reson for it is lasso is still based on linear property, which the data features are not really like even though it has been shrinkage.
